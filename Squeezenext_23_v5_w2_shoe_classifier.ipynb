```
train_datagen = ImageDataGenerator(rescale=1. / 255)
test_datagen = ImageDataGenerator(rescale=1. / 255)

train_generator = train_datagen.flow_from_directory(
    directory = r'./gdrive/My Drive/MLAI_files/Miro-project-images/Dataset_actual/Train', 
    target_size=(224,224), 
    color_mode = 'rgb', 
    class_mode = 'categorical',
    shuffle = True,
    batch_size = 32)

valid_generator = test_datagen.flow_from_directory(
    directory = r'./gdrive/My Drive/MLAI_files/Miro-project-images/Dataset_actual/Test', 
    target_size=(224,224), 
    color_mode = 'rgb', 
    class_mode = 'categorical',
    shuffle = True,
    batch_size = 32)
```

```
STEP_SIZE_TRAIN = train_generator.n//train_generator.batch_size
STEP_SIZE_VALID = valid_generator.n//valid_generator.batch_size
```

```
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```
def my_gen(gen):
    while True:
        try:
            data, labels = next(gen)
            yield data, labels
        except:
            pass
```

```
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```
checkpoint = ModelCheckpoint('./gdrive/My Drive/MLAI_files/Miro-project-images/Full_Dataset/squeezenext_weights.{epoch:02d}-{val_loss:.2f}.hdf5',
                             monitor='val_loss', save_best_only=True, mode='min')
```

```
callbacks_list = [checkpoint]
```

```
hist = new_model.fit_generator(generator = my_gen(train_generator),
                              steps_per_epoch = STEP_SIZE_TRAIN,
                              validation_data = my_gen(valid_generator),
                              validation_steps = STEP_SIZE_VALID,
                              epochs = 3,
                              callbacks = callbacks_list)
```
